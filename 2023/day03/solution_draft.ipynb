```python
data = open('puzzle.data').read().splitlines()


import re
numbers = []
for y, line in enumerate(data):
    x = 0
    while x < len(line):
        match = re.search('\d+', line[x:])
        if not match:
            break
    
        numbers.append( (int(line[match.start() + x: match.end() + x]), match.start() + x , match.end() + x - 1, y))
        x += match.end()

numbers
```

```python
marked_numbers = set()
for y, line in enumerate(data):
    x = 0
    while x < len(line):
        match = re.search(r'[^\d.]{1}', line[x:])
        if not match:
            break
    
        for number, x1, x2, y_ in numbers:
            if x + match.start() >= x1 - 1 and x + match.start() <= x2 + 1 and y in (y_ - 1, y_, y_ + 1):
                marked_numbers.add( (number, x1, x2, y) )
        x += match.end()

marked_numbers
marked_numbers
sum(number for number, *_ in marked_numbers)
    
```

```python
gear_ratio = 0
for y, line in enumerate(data):
    x = 0
    while x < len(line):
        match = re.search(r'[*]{1}', line[x:])
        if not match:
            break
    
        marked_numbers = set()
        for number, x1, x2, y_ in numbers:
            if x + match.start() >= x1 - 1 and x + match.start() <= x2 + 1 and y in (y_ - 1, y_, y_ + 1):
                marked_numbers.add( (number, x1, x2, y) )
        if len(marked_numbers) > 1:
            assert len(marked_numbers) == 2
            gears = tuple(marked_numbers)
            gear_ratio += gears[0][0] * gears[1][0]
        x += match.end()
gear_ratio
```
