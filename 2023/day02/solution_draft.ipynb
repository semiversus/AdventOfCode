```python
file_content = open('puzzle.data', 'r').read().splitlines()


fail = 0
for line in file_content:
    game, sets = line.split(':')
    _, id = game.split(' ')
    sets = [s.strip() for s in sets.split(';')]
    
    good_set = 0
    for set in sets:
        for cube in [s.strip() for s in set.split(',')]:
            number, color = cube.split(' ')
            number = int(number)
            if color == 'red' and number > 12:
                break
            if color == 'green' and number > 13:
                break
            if color == 'blue' and number > 14:
                break
        else:
            good_set += 1
    if good_set == len(sets):
        fail += int(id)

fail
```
